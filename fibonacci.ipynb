## continued fraction, e

```python
def con_frac_e(n):
    frac = 0
    for i in range(n, 0, -1):
        if i % 3 == 2:
            k = 2 * (i // 3 + 1)
        else:
            k = 1
        frac = 1 / (k + frac)
    return 2 + frac
```

```python
es = [con_frac_e(i) for i in range(20)]
es.insert(0, np.e)
x = range(-1, 20)
plt.rcParams['xtick.color'] = 'orange'
plt.rcParams['ytick.color'] = 'orange'
plt.ylim(np.e - 1, np.e + 1)
plt.plot(x, es)

```
